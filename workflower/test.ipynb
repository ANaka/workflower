```python
import redun
from redun import task, Scheduler
from scipy import stats
import numpy as np
import requests
import json
from bs4 import BeautifulSoup
import logging, time
from wiki_word_count import main
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
sns.set(context='talk', style='darkgrid')
```

```python
scheduler = Scheduler()
scheduler.logger.setLevel(logging.INFO)
```

```python
result = scheduler.run(main(n_articles=1000))
```

```python
df = pd.DataFrame(result)
```
